# Random Walk
A random walk, known as a stochastic or random process, is a mathematical tool that describes a path consisting of a succession of random steps in some mathematical space, such as the integers.

Random walks have applications to engineering and many scientific fields, including ecology, psychology, computer science, physics, chemistry, biology, economics, and sociology. It explains the observed behaviors of many processes in the fields mentioned above and thus serves as a fundamental model for the recorded stochastic activity.
For example, the path taken by a molecule when moving in gas or liquid, the path taken by a herbivore, stock price fluctuations, and the financial condition of a gambler; These are items that can be modeled by lagging.

Karl Pearson first introduced the term random walk in 1905.

# Classical Walk

Classical walk refers to a random process in which one is randomly moved (with equal or different probabilities) to the left or right. This is an example of a one-dimensional random walk which moving in a straight line. This movement to the right or left can be simulated by throwing a coin. In such a way that each step is a throw. In this case, the person moves in discrete points and we do not have a continuous movement. Now we want to find the position of the person after n steps. To do this, we set the starting point to zero and call the particle position after n steps m. Suppose we have  steps to the right and   steps to the left. So, we will have:

\begin{equation}
    n_{R} +n_{l} =n \\ n_{R} -n_{l} =m. \\
    \stackrel{}{\longrightarrow}{\rm \; \; \; }n_{R} =\frac{n+m}{2} ,{\rm \; \; }n_{l} =\frac{n+m}{2} 
\end{equation}

Suppose, the probability of moving to the right is p and the probability of moving to the left is q. The probability of the person being at m point after n step is as follows:

\begin{equation}
P_{n} (m)=\left(\begin{array}{l} {n} \\ {n_{R} } \end{array}\right)p^{n_{R} } q^{n-n_{R} } 
\end{equation}

A specific example is that the probability of p and q are equal, in this case we have:

\begin{equation}
P_{n} (m)=\frac{n!}{n_{R} !(n-n_{R} )!} \left(\frac{1}{2} \right)^{n}
\end{equation}

Now the goal is calculate of X${}_{rms}$ $x_{r.m.s} =\sqrt{\overline{m^{2} }-\overline{m}^{2} } $in wich $\overline{m}=2\overline{n_{R} }-n$, so:

\begin{equation}
\overline{n_{R} }=\sum \frac{n!}{n_{R} !(n-n_{R} )!} n_{R} p^{n_{R} } q^{n-n_{R} }   \\ {{\rm \; \; \; \; }=p\frac{\partial }{\partial p} \left(\sum _{n_{R} =0}^{\infty }\left(\begin{array}{l} {n} \\ {n_{R} } \end{array}\right)p^{n_{R} } q^{n-n_{R} }  \right)} \\ {{\rm \; \; \; \; }=p\frac{\partial }{\partial p} (p+q)^{N} =np}
\end{equation}

It explicitly becomes $\overline{m}=0$. We obtain $\overline{m^{2} }$ with a similar approach, the first obtain $\overline{n_{R} {}^{2} }$:

\begin{equation}
\overline{n_{R} {}^{2} }=p\frac{\partial }{\partial p} \left(p\frac{\partial }{\partial p} (p+q)^{n} \right)=np+np^{2} (n-1)
\end{equation}

With straightforward calculations the relation  $\overline{m^{2} }=n$ would be obtained. Then we will see that 

\begin{equation}
x_{r.m.s.} =n^{1/2},
\end{equation}


# Quatum Walk

A vector introduces a quantum state in the Hilbert space, and a unitary operator applies the evolution of the non-interacting systems. The quantum analogs of classical random walks are called Quantum walks.

The notion of a quantum random walk was originally discovered by Aharonov et al in 1993 [32]. Contrary to the classical random walk where the walker has an equal probability of making a step of a given length in a given direction (for one dimension this is left and right, two dimensions would also includes upwards and downwards, etc), quantum walks are based upon probability amplitudes. 
Quantum walks are quantum analogues of classical random walks. In contrast to the classical random walk, where the walker occupies definite states and the randomness arises due to stochastic transitions between states, in quantum walks randomness arises through: (1) quantum superposition of states, (2) non-random, reversible unitary evolution and (3) collapse of the wave function due to state measurements. 

Quantum analogues of random walk provide speed-up in computational power due to the quantum features such as quantum superposition, interference and entanglement for various algorithms such as element distinctness, spatial search, graph connectivity, etc and spreads quadratically faster in position space. This notion brings up a motivation to introduce algorithms for quantum computers that solve the problem exponentially faster than the best classical algorithm. [Quantum Information Spreading in a Disordered QuantumWalk]  

Quantum walk has different types and is generally divided into two categories: Discrete Time Quantum Walk Model (DTQWM) and Continues Time Quantum Walk Model (CTQWM).

In the discrete model, walks are assumed to be discrete, that is, like moving on a straight line that can only be placed at unique points on the line. The evolution of a quantum walk in discrete time is specified by the product of two unitary operators: (1) a "coin flip" operator and (2) a conditional shift operator, which are applied repeatedly.

Continuous-time quantum walks arise when one replaces the continuum spatial domain in the Schrödinger equation with a discrete set.Under particular conditions, continuous-time quantum walks can provide a model for universal quantum computation.
For the first time, Farhi and Guttman used continuous quantum walk algorithm to navigate the decision tree in their research [2]. The origin of this algorithm goes back to the work of Feynman et al., In which the unitary operator is defined as, 

\begin{equation}
    U = e^{-iHt}
\end{equation}

in which  is Hamiltonian operator. And the final position of the particle at time t is given as follows:

\begin{equation}
    i\hbar \frac{\partial \ket{\Psi(t)}}{\partial t} = H \ket{\Psi (t)} \\
    \ket{\Psi(t)} = e^{-iHt}\ket{\Psi (0)}
\end{equation}

In this walk, the walk space is the same as Hilbert space.
The continuous time quantum walk has some notable contrasts with the discrete random walk which will be described in this section, namely: 1- CTQW are generally described by Markov processes 2- CTQW evolves under a Hamiltonian which is deﬁned with respect to a graph 3- No coin operator is required in a CTQW. 


# Discrete Time Quantum walk

Discrete time quantum walk is classified into two types: coin walk and scatter walk. In coin quantum walk, the quantum particle is placed on the vertices of the graph, and another degree of freedom called "coin space" is used to unify the particle dynamics [70] and [71]. 
Dispersion quantum walk is another type of discrete quantum walk that arises from the idea of  scatter particle. This quantum walk was first introduced by Hillery et al. [72] in which a quantum particle is placed on the edges of a graph. In this type of quantum walk, the coin space is unnecessary and each edge describes two distinct and normal orthogonal states. One corresponding to the presence of a particle on the edge and moving in a certain direction and the other describing the presence of a particle on the same edge and move in the opposite direction. In other words, if the two vertices a and b are connected by an edge, this edge represents the two states ,ab moving from vertex ,ba. 
This structure allows us to study quantum walk from the perspective of dispersion theory. The general properties of this type of quantum walk can be studied from the scattering matrix, or in other words, the transfer and reflection functions (which are the characteristics of a graph). This approach has major advantages, including the possibility of generalizing the loop to any complex and large graph by combining smaller graphs. 
 
 ## Coined quantum walks
In coin walk, the position of the system is introduced by $\ket{n}$ in a Hilbert space $H_p$ where $\{\ket{n}\in z\}$. The direction of the movements is defined in another Hilbert space called $H_c$ (coin space). The coin space can be considered as a spin $1/2$ system which the movements to left and right can be introduced by $\ket{\frac{1}{2},\frac{-1}{2}}=\ket{L}$, and $\ket{\frac{1}{2},\frac{1}{2}}=\ket{R}$.According to these definitions, our particle will move to the right whenever the state of the coin is $\ket{R}$, and vice versa for state $\ket{L}$. So, a one-dimensional quantum walk system can be introduced by the tensor product of two $H_p$ and $H_c$.

\begin{equation}
    H = H_c \otimes H_p
\end{equation}

The product of two unitary operators specifies the evolution of a quantum walk in discrete time: (1) a "coin flip" operator and (2) a conditional shift operator, which is applied repeatedly. When the conditional shift operator applies to the system, the result can be shown by

\begin{equation}
    S\ket{R}\ket{n} = \ket{R}\ket{n+1} \\
    S\ket{L}\ket{n} = \ket{L}\ket{n-1}
\end{equation}

We can introduce the linear operator $S$ as

\begin{equation}
    S = \ket{R}\bra{R} \otimes \Sigma_{n=-\infty}^{\infty}\ket{n+1}\bra{n} + \ket{L}\bra{L} \otimes \Sigma_{n=-\infty}^{\infty}\ket{n-1}\bra{n}
\end{equation}

A popular choice for coin operator is the Hadamard gate, which, with respect to the standard z-component spin basis, has matrix representation

\begin{equation}
    H=\frac{1}{\sqrt{2}}\mymatrix{rr}{1 & 1 \\ 1 & -1 }
\end{equation}

Applying Hadamard on coin space is the first step and then $S$ will apply to shift the particle 

\begin{equation}
    (H \otimes I)\ket{00} = \frac{\ket{00} + \ket{10}}{\sqrt{2}}\rightarrow^S\rightarrow \frac{\ket{01} + \ket{1-1}}{\sqrt{2}}
\end{equation}

Where in $\ket{ab}$, $a$ shows the state of the coin (0 cause right-shift and $1$ left-shift), and $b$ introduce the location of the particle. Based on this definition, when particle start from $n = 0$, it will shift to $n = 1$ or $n = -1$ with $50%$ probability according to state of the coin. If we measure the particle after each shift, the results will be the same as the classic case since the quantum nature of the system after measurement collapse to the classical regime. To find the results based on the quantum nature of the system, we should continue to the next shifts without any measurements. 
The correlation between different position states makes some Constructive or destructive interference that shows the quantum nature of the calculations. In the quantum walk, the distribution is not normal, and the standard deviation is not equal to $\sqrt{t}$.
Accordingly, we can introduce the shift operator as

\begin{equation}
    u = S(H \otimes I)
\end{equation}

For example, with some calculation, it can be shown that the probability of presence of particle at point 1 is as follows

\begin{equation}
\ket{\psi (1)} =\frac{1}{\sqrt{2} } (\ket{0} \otimes \ket{1} + \ket{1} \otimes \ket{-1} ),
\end{equation}

and so for the second and third step we will have

\begin{equation}
\ket{\psi (2)} =\frac{1}{2} (\ket{1} \otimes \ket{2} +(\ket{0} + \ket{1} )- \ket{1} \otimes \ket{-2} ) \\ 
\ket{\psi (3)} =\frac{1}{2\sqrt{2} } (\ket{0} \otimes \ket{3} +(2 \ket{0} + \ket{1} ) \otimes \ket{1} - \ket{0} \otimes \ket{-1} + \ket{1} \otimes \ket{-3} ).
\end{equation}



After $t$ steps, the state of system can be introduced by $\ket{\psi (t)} = u^t \ket{\psi (0)}$, and generally can be considered as

\begin{equation}
    \ket{\psi (t)} = \Sigma_{n = -\infty}^{\infty} (A_n (t)\ket{0} + B_n (t) \ket{1}) \ket{n}
\end{equation}

<img src="002.PNG" alt="Drawing" style="width: 600px;height: 300px"/>
Table.2) The probability of position of the particle after t times in quantum walk.



To normalize

\begin{equation}
    \Sigma_{n = -\infty}^{\infty} |A_n (t)|^2 + |B_n (t)|^2 = 1
\end{equation}

Recursive relationships can calculate the coefficients $A_n$ and $B_n$.

\begin{equation}
    \ket{\psi (t+1)} = S(H \otimes I) \psi (t) \\
    \\
    \ket{\psi (t+1)} = \Sigma_{n = -\infty}^{\infty} \frac{A_n (t) + B_n (t)}{\sqrt{2}} \ket{0} \otimes \ket{n+1} + \frac{A_n (t) - B_n (t)}{\sqrt{2}} \ket{1} \otimes \ket{n-1} 
\end{equation}

These equations lead us to

\begin{equation}
  A_n (t+1) = \frac{A_{n-1}(t) + B_{n+1} (t)}{\sqrt{2}} \\
  B_n (t+1) = \frac{A_{n+1}(t) - B_{n+1} (t)}{\sqrt{2}} 
\end{equation}

Considering initial conditions $A_n (0) = 1$ for $n=0$ ( and $A_n (0) = 0$ for $n \neq 0$) and $B_n (0) = 0$. 
The probability distribution is

\begin{equation}
    p(t,n) = |A_n (t)|^2 + |B_n (t)|^2.
\end{equation}

Unlike the classic system, the probability distribution of the quantum walk is an asymmetric function, and the maximum probability is not around the initial central position. By changing the initial state, the results will be changed (Fig. 1 and 2) since the Hadamard gate has different effects on states $\ket{0}$ and $\ket{1}$. 

\begin{equation}
    H\ket{0} = \frac{1}{\sqrt{2}} (\ket{0} + \ket{1})\\
    H\ket{1} = \frac{1}{\sqrt{2}} (\ket{0} - \ket{1})
\end{equation}

To get the symmetric distribution, we can introduce different operators as coin operators. For example, we can use the $Y$ operator defined by

\begin{equation}
    Y = \frac{1}{\sqrt{2}} \mymatrix{rr}{1 & i \\ i & 1 }
\end{equation}

Also, we can change the initial state as

\begin{equation}
    \ket{\psi(0)} = \frac{1}{\sqrt{2}} (\ket{0} - i \ket{1})
\end{equation}

to get the symmetric distribution by applying Hadamard coin. The symmetric probability distribution of the system after $100$ times applying the Hadamard operator to initial state $\ket{\psi(0)} = \frac{1}{\sqrt{2}} (\ket{0} - i \ket{1})$ is shown in Fig. 3.

## Quantum walk code

### For 1 step n=2

```python
#discrete time quantum random walk with a hadamard coin
n=2
steps=1

qpos = QuantumRegister(n,'qpos')
qcoin = QuantumRegister(1,'qcoin')
cpos = ClassicalRegister(n,'cr')
QC = QuantumCircuit(qpos, qcoin)


#initialized
#QC.x(qcoin[0])

for i in range(steps):

#coin
    QC.h(qcoin[0])
    QC.barrier()

#increment gate operator
    for i in range(n):
        QC.mct([qcoin[0]]+qpos[i+1:], qpos[i], None, mode='noancilla')
    QC.barrier()

#decrement gate operator
  
    QC.x(qcoin[0])
    for i in range(n):
        if i+1 < n: QC.x(qpos[i+1:])
        QC.mct([qcoin[0]]+qpos[i+1:], qpos[i], None, mode='noancilla')
        if i+1 < n: QC.x(qpos[i+1:])
    QC.barrier()
a=n/2
p=math.floor(a)
#print(p)
for k in range(n):
    if(k<p):
        QC.swap(qpos[n-1-k],qpos[k])

#print(a)
    
display(QC.draw(output="mpl"))
```

```python
QC.measure_all()

    # # Execute the circuit on the qasm simulator
    
job = execute(QC,Aer.get_backend('aer_simulator'),shots=10000)
counts = job.result().get_counts(QC)   
print(counts)

plot_histogram(counts, color='midnightblue', title="New Histogram")

#aer_sim = Aer.get_backend('aer_simulator')
#qobj = assemble(had)
#result = aer_sim.run(qobj).result()
#counts = result.get_counts()
#plot_histogram(counts)
```

## Decision Making 

Decision-making models are central in economics and psychology. Finding suitable models which can approach human decision-making behavior is, as expected, a very challenging task. Economics has for a long time assumed models which are based on a particular axiomatic skeleton. The axioms are “reasonable” approximations of general decision situations. 
A central starting point in preference modeling in economics consists in proving that there exists an equivalence between the preference relation of an object x over an object $y$, denoted as $x > y$, if and only if there exists a utility function $u(.)$, which maps a set of objects into $R$, such that $u(x) > u(y)$. This is not an easy task. However, as any good microeconomic theory textbook will show for a finite set of objects $X$ (to which $x$ and $y$ belong), the equivalence is relatively easy to show [[D01]](https://doi.org/10.1017/CBO9781139003261). 
In psychology, Decision-Making is regarded as the cognitive process resulting in selecting a belief or a method of action among several possible alternative options. It could be either rational or irrational. Decision-making process is a reasoning process based on assumptions of the decision-maker's values, preferences, and beliefs. Every decision-making process produces a final choice, which may or may not prompt action [[D02]](https://en.wikipedia.org/wiki/Decision-making).
One famous example of decision-making theory is the prison's dilemma. One famous example of decision-making theory is the prison's dilemma. A prisoner's dilemma is a situation where individual decision-makers always have an incentive to choose to create a less than optimal outcome for the individuals as a group.

## Prison's Dilemma
The prisoner's dilemma is a standard example of a game analyzed in game theory that shows why two completely rational individuals might not cooperate, even if it appears that it is in their best interests to do so. It was originally framed by Merrill Flood and Melvin Dresher while working at RAND in 1950. Albert W. Tucker formalized the game with prison sentence rewards and named it "prisoner's dilemma" [[D03]](https://en.wikipedia.org/wiki/Prisoner%27s_dilemma).
The game is presented as:
Two members of a criminal organization are arrested and imprisoned. Each prisoner is in solitary confinement without communicating with the other. The prosecutors lack sufficient evidence to convict the pair on the principal charge, but they have enough to convict both on a lesser charge. Simultaneously, the prosecutors offer each prisoner a bargain. Each prisoner can either betray the other by testifying that the other committed the crime or cooperate with the other by remaining silent. The possible outcomes are:
* If A and B betray the other, they serve two years in prison,
* If A betrays B, but B remains silent, A will be set free, and B will serve three years in prison,
* If A remains silent but B betrays A, A will serve three years in prison, and B will be set free,
* If A and B both remain silent, both of them will serve only one year in prison (on the lesser charge).

It is implied that the prisoners will have no opportunity to reward or punish their partner other than the prison sentences they get and that their decision by itself will not affect their reputation in the future. As betraying a partner offers a greater reward than cooperating with them, all purely rational self-interested prisoners will betray the other, meaning the only possible outcome for two purely rational prisoners is for them to betray each other, even though cooperation would yield a greater reward.

This section will show how a Prison's dilemma's game can be modeled based on a quantum walk algorithm.

## Prison's Dilemma as a quantum walk model
Iterated bipartite quantum games can be implemented in the discrete-time quantum walk on the line. This section studies a discrete-time quantum walk on a line with two particles defined as two agents. Classically, random walks with $K$ particles are equivalent to $K$ independent single-particle random walks. In the quantum case, though, a walk with $K$ particles may contain quantum correlation, thus offering a resource unavailable in the classical scenario, introducing exciting features. Also, in the case of identical particles, we have to consider the effects of quantum statistics, giving an additional feature to quantum walks that can also be exploited. 
We model the game based on two rational agents chosen from a restricted set of two-qubit unitary operations. 
In the following, we present a quantum version of the Prisoner’s Dilemma in which both players use mixed strategies as a specific example [[D04]](https://arxiv.org/abs/quant-ph/0607143). 

### Discrete-time quantum walk
The Hilbert space of a quantum walk on a line comprises two parts, $H = H_x \otimes H_c$.
$H_c$ is spanned by two orthonormal states $\ket{0_i}, \ket{1_i}$ as “coin” subspace. The spatial subspace, $H_x$, is traversed by the orthonormal set of position eigenstates, $|\ket{x_i}$, with $x ∈ Z$ labeling discrete sites on a line and we use this symbol to distinguish the position eigenstates from coin space. Repeated application $H_c$ generates the evolution. The evolution is generated by repeated application of a composite unitary operator $U$ which implements a coin operation, followed by a conditional shift in the walker's position.
We model the Prison's dilemma as a game where two walkers play, and two coins can model their decision in each step.
The quantum walk with two walkers A,B takes place in a Hilbert space $H_{AB} = H_A \otimes H_B$, where $H_A$ and $H_B$. After $N$ steps, a pure state characterized by a density operator $\rho_0 = \ket{\psi (0)}\bra{\psi (0)}$ evolves to $\rho_N = U^N \rho_0 U^{\dagger N}$ with [[D05]](http://www.phys-info.org/uploads/3/8/1/3/3813936/physreva.74.042304.pdf)

\begin{equation}
    U = S(I \otimes U_c)
\end{equation}

where $U_c$ is a unitary operation in coin subspace, $I$ is the identity and $S$ is a shift operation in $H_{AB}$.

## Model Prison's dilemma by quantum walk
We can model the game based on the quantum walk algorithm. In this model, two players are considered two walkers, and their decision is made by tossing a coin for each player. We can see a graphic example of Prison's dilemma in the following figure

As the figure shows, the player's decision is considered coin states where each decision confessor or being silent is considered getting Heads or Tails. When the prison decides to remain silent or confess, it will be coded as $\ket{0}$ or $\ket{1}$ respectively.
On the other hand, the years of conviction are coded by position states. Since we have four states for each prison, we need two qubits for each one. So, we have two qubits for two coins presenting the decision of each player and four qubits in position space. At first, we consider the decision coins are Hadamard coins where are presented by $U(\pi/2, 0, \pi)$, and each player's state initially is $\ket{00}$. 

Now, we are going to simulate the system:

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, execute, Aer,assemble, transpile
from qiskit import BasicAer
from random import randrange
from qiskit.tools.visualization import plot_histogram, plot_state_city
import numpy as np
from qiskit.quantum_info.operators import Operator, Pauli
from qiskit.quantum_info import Operator, Statevector
```

We need six qubits. The qubits $q[0]$ and $q[1]$ are presenting each player's decision. We first consider a Hadamard coin which shows an equal probability for each coin state.

```python
q = QuantumRegister(6)  # "Qubits q[0] and q[1] for coin 1 and 2", "q[2] and q[3] for player 1" and q[4] and q[5] for player 2" 
c = ClassicalRegister(6)
Q = QuantumCircuit(q,c)

#Coin Operators
# Apply u gate as Hadamard gate to each coin
Q.u(np.pi/2,0,np.pi,q[0])   
Q.u(np.pi/2,0,np.pi,q[1])
```

Initially, we suppose two players are in their $\ket{00}$ state. It means both of the prisons are considered free. We show the consequence of their decision to confess or be silent coming from coin space.

```python
#S operator

Q.x(q[0])
Q.x(q[1])
Q.mct([q[0],q[1]],q[3])
Q.mct([q[0],q[1]],q[5])
Q.x(q[0])
Q.x(q[1])
Q.barrier()


Q.x(q[0])
Q.mct([q[0],q[1]],q[2])
Q.mct([q[0],q[1]],q[3])
Q.x(q[0])
Q.barrier()

Q.x(q[1])
Q.mct([q[0],q[1]],q[4])
Q.mct([q[0],q[1]],q[5])
Q.x(q[1])
Q.barrier()

Q.mct([q[0],q[1]],q[2])
Q.mct([q[0],q[1]],q[4])




Q.measure_all()

Q.draw(output='mpl')
```

Now we can measure each qubit.

```python
# Execute the circuit on the qasm simulator
# Use Aer's qasm_simulator
backend_sim = Aer.get_backend('qasm_simulator')

# Execute the circuit on the qasm simulator.
# We've set the number of repeats of the circuit
# to be 1024, which is the default.
job_sim = execute(Q, backend_sim, shots=1000)

# Grab the results from the job.
result_sim = job_sim.result()
counts = result_sim.get_counts(Q)
print(counts)
    


plot_histogram(counts, color='midnightblue', title="Decision Histogram")
```
