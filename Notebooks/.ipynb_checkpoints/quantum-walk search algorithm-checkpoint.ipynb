# Quantum Walk Search Algorithm

The Graver algorithm can search for a specific target in a disordered database by increasing the square speed compared to classical algorithms. But the Graver algorithm cannot analyze all the problems, and, in some cases, it is even slower than the classical algorithms. Quantum walk has been successfully applied to several quantum algorithms ([[S01]](https://arxiv.org/abs/quant-ph/0311001), [[S02]](https://ieeexplore.ieee.org/document/1366222), and [[S03]](https://arxiv.org/abs/quant-ph/0310134). An abstract search algorithm is also proposed to find an actual vertex on any regular and desired graph by correcting the standard spin coin operator [[S04]](https://arxiv.org/abs/quant-ph/0402107).
For the first time Shenvi et al. proposed a version of a quantum search algorithm based on coin quantum spin on an n-dimensional super cube [[S05]](https://journals.aps.org/pra/abstract/10.1103/PhysRevA.67.052307). Until now, we will know this algorithm by the abbreviation SKW (Shenvi, Kempe and Whaley). The proposed algorithm is designed to find one (or more) definite vertices (marked by an oracle) in a cube grid, and the number of oracles calls to find the marked vertex is similar to the Graver algorithm.
Despite this similarity, the SKW algorithm is important because it is not possible to efficiently implement the diffusion operator in the Graver algorithm (because this operator in the Graver algorithm has $2^n \otimes 2^n$ dimension), in the SKW algorithm, this operator has the dimension $n × n$, and it is possible to implement it more easily [[S06]](https://journals.aps.org/pra/abstract/10.1103/PhysRevA.81.022308). In the following section, we will give a full description of this algorithm.

## Comparison of quantum walk with Grover algorithm, differences and similarities

The two search algorithms by quantum spin method and gravure algorithm (despite their different nature) also have differences and similarities. In this section, we will examine these differences and similarities.
Similarities of the two algorithms:

1.	Both algorithms start with an equal overlap of all possible states.
2.	In both algorithms, we use the gravure diffusion operator ($G$).
3.	Both algorithms can be considered as a period in two-dimensional space.
4.	Both use an oracle to give phase -1 to the answer and thus mark the target.
5.	Both need time of order $O(\sqrt{N})$ to reach the answer.
6.	In both algorithms, we measure the state of the system at a specific time to find the vertex of the target with the highest possible probability.

We will now explain these differences and their effect on the performance and implementation of the algorithm.
Recall that the Graver algorithm is mapped to exactly one period in two-dimensional space. Each query of the Graver algorithm will be equivalent to one period in this subspace. In the quantum walk algorithm, this algorithm can also be considered as a two-dimensional period, but there are two main differences:

1.	Unlike the Graver algorithm, which is an accurate time-mapping in two-dimensional subspace, the quantum walk algorithm is an approximate two-dimensional mapping.
2.	The two-dimensional subspace in which the spin algorithm occurs is swept by the vectors $\ket{\Psi_1}$ and $\ket{\Psi_0}$, not by the vectors $\ket{\Psi_0}$ and $\ket{0}$. Thus, the final state of the algorithm, precisely and purely, does not only include the target item, but also a small distribution of the heads of the nearest and farthest neighbors.
3.	The third difference is in how the Graver operator is used during the algorithm. In the Graver algorithm, this operator is applied to the entire   dimensional search space (corresponding to the vertex space in the spin algorithm). While in the quantum walk algorithm, the effect of the Graver operator as a coin operator is limited to the edge space (coin space). Since most implementations of quantum computers involve multiple types of qubits, which have different sets of gates, this fact can be very useful in the physical implementation of the algorithm. We can use this point and select the coin space with qubits that are suitable for implementing the Graver diffuser [[S05]](https://journals.aps.org/pra/abstract/10.1103/PhysRevA.67.052307).
